## The Law of Large Numbers  
1. **The Strong Law.**  
Sample $N = 10,000$ random numbers $X_1, \dots, X_N$ exponentially distributed with parameter 1. Plot the values for $N= 1, \dots, 10,000$  
2. **The weak law.** Define a function that will return the empirical mean of a size N as above. Plot the histograms (PDF and CDF) of a sample size 10,000 of the empirical mean for $N=100$ and $N= 10,000$.

````python
sample_size = [10000, 100]

def empirical_mean(size_of_sample):
    '''
    input: 
    size_of_sample: (int) determine the sample size that will be generated from exponential distribution  

    output:  
    data: (array) data generated by the function 
    empirical_mean: (float) the mean of the sample
    '''
    data = np.random.exponential(scale = 1, size = size_of_sample)
    data_mean = np.sum(data)/size_of_sample
    return data, data_mean

data_1, mean_1 = empirical_mean(sample_size[0])
data_2, mean_2 = empirical_mean(sample_size[1])

print(f'The empirical mean of sample size {sample_size[0]} is {mean_1} and for sample size {sample_size[1]} is {mean_2}')
print('From the result above, we can conclude that the bigger the sample size, the closest the empirical mean to the real mean')
````

````python
plt.hist(x = data_1, bins = 50, density = True)
plt.hist(x = data_2, bins = 50, density = True)
plt.title(f'PDF of Exponential Distibution of Sample Size {sample_size[0]} and {sample_size[1]}')
plt.legend(sample_size)
plt.show()
````

````python
plt.hist(x= data_1, bins = 50, density= True, cumulative= True)
plt.hist(x = data_2, bins = 50, density= True, cumulative= True)
plt.title(f'CDF of Exponential Distibution of Sample Size {sample_size[0]} and {sample_size[1]}')
plt.legend(sample_size)
plt.show()
````

# The Central Limit Theorem
The approximation (1.5) of the expectation in terms
of the empirical mean is not exact for finite $N$. The error is controlled by the central limit theorem. For a sample $X_1, \dots , X_N$ of the random variable $X$ of mean $E[X]$ and variance $\sigma^2$ this theorem says that the sum $S_N \approx X_1 + ⋯ + X_N$ behaves like
\begin{equation}
S_N \approx NE[X] + \sqrt{N} \sigma Z
\end{equation}
where $Z$ is a standard Gaussian random variable. More precisely, this means that
\begin{equation}\label{clt}
\lim_{N\to \infty} \frac{S_N - NE[X]}{\sigma \sqrt{N}}=Z
\end{equation}
The limit should be understood here as the convergence in distribution. Practically speaking, this means that the histogram of the random variable
$\frac{S_N - NE[X]}{\sigma \sqrt{N}}$
should resemble the one of a standard Gaussian variable when $N$ is large. We check this numerically.
\begin{enumerate}
\item Let $S_N = X_1 + \dots + X_N$ where the $X_i$ are exponentially distributed random variables of parameter 1. Define a function in Python using the command
```def`` that returns for a given $N$ the value $Y_N = \frac{S_N - NE[X]}{\sigma \sqrt{N}}$ 
\item Plot the histograms (PDF) of a sample of size 10,000 of $Y_N$ for $N$ = 100. What do you notice?
\item Compare the above to the histogram of a sample of size 10,000 of points gen-
erated using the standard Gaussian distribution.
\end{enumerate}

````python
def clt(N):
    ```
    input:
    N : integer, number of random variable
    output:
    X : random variables 
    Y_N : CLT 
    ```
    X = np.random.exponential(scale = 1)
    return X, (np.sum(X) - N*np.mean(X))/((np.var(X)*N)**0.5)
````
